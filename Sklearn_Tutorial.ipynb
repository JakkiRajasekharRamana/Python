```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
# simple plot to understand the data

plt.figure(figsize =(10, 10))
plt.bar(games['Genre'], games['Global_Sales'])
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.show()
```

```python
games.isnull().sum()
```

```python
#filling all the nul values

games.fillna(method="bfill", inplace=True)
games.isnull().sum()
```

```python
# label encoding, i.e., converting all object data types into numerical data

for col in games.columns:
    if games[col].dtype == 'object':
        games[col] = LabelEncoder().fit_transform(games[col])
```

```python
from sklearn.metrics import *
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# creating the linear regression model from the sklearn library
# note that the sklearn linear regression model uses the inverse matrix approach rather than the gradient descent approach as it is more efficient

linear = LinearRegression()
```

```python
# xdf is the input variables while ydf is the target variable, i.e., variable which we are going to predict (global sales)

xdf = games.drop(columns = ['Global_Sales'])
ydf = games["Global_Sales"]
```

```python
xdf.head(5)
```

```python
ydf.head(5)
```

```python
# splitting the data into training and testing sets in the 80-20 ratio

x_train, x_test, y_train, y_test = train_test_split(xdf, ydf, test_size = 0.2, random_state = 20)
```

```python
# This normalization step has not been included as it is creating an inaccurate model

'''from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler(feature_range = (0, 1))
x_train = mms.fit_transform(x_train)
x_test = mms.fit_transform(x_test)
x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)'''
```

```python
# training the model using training set

linear.fit(x_train, y_train)
```

```python
# finding the score of the model out of 100 using the testing set

r2_score = linear.score(x_test, y_test)
```
